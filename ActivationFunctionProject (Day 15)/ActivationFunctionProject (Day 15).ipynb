```python
from matplotlib import pyplot
def relu(x):
    return max(0.0,x)

inp = [x for x in range(-10,10)]
outp = [relu(x) for x in inp]
pyplot.plot(inp,outp)
pyplot.show()
```

```python
from math import exp
from matplotlib import pyplot

def sigmoid(x):
    return 1.0/(1.0 + exp(-x))

inp = [x for x in range(-10,10)]
outp = [sigmoid(x) for x in inp]
pyplot.plot(inp,outp)
pyplot.show()
```

```python
def tanh(x):
    return (exp(x) - exp(-x))/(exp(x) + exp(-x))

inp = [x for x in range(-10,10)]
outp = [tanh(x) for x in inp]
pyplot.plot(inp,outp)
pyplot.show()
```

```python
def linear(x):
    return x

inp = [x for x in range(-10,10)]
outp = [linear(x) for x in inp]
pyplot.plot(inp,outp)
pyplot.show()
```
